```python
with open('the-verdict.txt', 'r' , encoding='utf-8') as f:
    raw_text = f.read()
print('total number of characters:', len(raw_text))
print(raw_text[:99])
```

```python
import re 
text = "hello, world. This, is a test message "
result = re.split(r'([,.]|\s)',text) 
#here we are splitting the text by comma, dot and space. The parentheses around the regex pattern means that the delimiters will also be included in the resulting list.
print(result)

result = [item for item in result if item.strip() != '']
print(result)
```

```python

text2 = "hello, world . Is this-- a test ?"
result2 = re.split(r'([,.?]|\s|--)', text2)
result2 = [item for item in result2 if item.strip() != '']
print(result2)
```

```python
preprocessed_text = re.split(r'([.,:"!?;_()\']|--|\s)',raw_text)
preprocessed_text = [item for item in preprocessed_text if item.strip() != '']
print(preprocessed_text[:100])
print('total number of tokens:', len(preprocessed_text))
```

NOW WE WILL GIVE UNIQUE TOKEN IDS TO THE TOKENS CREATED 

```python
all_words = sorted(set(preprocessed_text))
vocab_size = len(all_words)
print('vocab size:', vocab_size)
```

```python
#now i will provide unique id to the tokens. This is called tokenization. We will create a dictionary that maps each unique token to a unique integer id.
vocab = {token: integer  for integer , token in enumerate(all_words)}
for i, item in enumerate(vocab.items()):
    print(item)
    if i >= 100:
        break
```
