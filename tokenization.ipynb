```python
preprocessed_text = re.split(r'([.,:"!?;_()\']|--|\s)',raw_text)
preprocessed_text = [item for item in preprocessed_text if item.strip() != '']
print(preprocessed_text[:100])
print('total number of tokens:', len(preprocessed_text))
```

NOW WE WILL GIVE UNIQUE TOKEN IDS TO THE TOKENS CREATED 

```python
all_words = sorted(set(preprocessed_text))
vocab_size = len(all_words)
print('vocab size:', vocab_size)
```

```python
#now i will provide unique id to the tokens. This is called tokenization. We will create a dictionary that maps each unique token to a unique integer id.
vocab = {token: integer  for integer , token in enumerate(all_words)}
for i, item in enumerate(vocab.items()):
    print(item)
    if i >= 100:
        break
```

CREATE A TOKENIZER CLASS 

```python
class simpleTokenizer:
    def __init__(self, vocab):
        self.str_to_int = vocab 
        self.int_to_str = {i : s for s, i in vocab.items()}
        #here we have s as token and i as integer. We are creating a reverse mapping from integer to token.
    def encode(self, text):
        preprocessed = re.split(r'([.,:"!?;_()\']|--|\s)', text)
        preprocessed = [item for item in preprocessed if item.strip() != '']
        ids = [self.str_to_int[s] for s in preprocessed]
        return ids
    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        #when we join the tokens then there is no space between symbol and word thats why re.sub is used 
        text = re.sub(r'\s+([.,"!?;()\'])', r'\1', text)
        return text
```

SPECIAL CONTEXT TOKENS USED WHEN WE HAVE SOME NEW WORDS ON WHICH TOKENS WERE NOT CREATED ,, two kind of tokens can be used 1 -> for unknown values 2-> for values ie end of text (if we have text from different sources and we want to saparate them then this is used ) 

```python
all_tokens = sorted(set(preprocessed_text))
all_tokens.extend(["<|endoftext|>","<|unk|>"])
#for every token there will be a unique integer id. The special tokens <|endoftext|> and <|unk|> are added to the vocabulary to represent the end of text and unknown tokens respectively.
vocab = {token:integer for integer , token in enumerate(all_tokens)}
#two tokens have been added to the vocabulary, so the vocab size will be 1130 + 2 = 1132
len(vocab.items())
```

```python
class SimpleTokenizerV2:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = { i:s for s,i in vocab.items()}
    
    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [item.strip() for item in preprocessed if item.strip()]
        preprocessed = [
            item if item in self.str_to_int 
            else "<|unk|>" for item in preprocessed
        ]

        ids = [self.str_to_int[s] for s in preprocessed]
        return ids
        
    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        # Replace spaces before the specified punctuations
        text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)
        return text
```

now we have to create two tensors one for the input text and other for the text that needs to be predicted               we have to do this in a structed format and make use of data loader                           1 tokenizing the entire text 
2 use a sliding window to chunk the book into overlapping sequence of max_length 
3 return the total number of rows in the dataset
4 return a single row from the dataset

```python
from torch.utils.data import Dataset, DataLoader    
class GPTDatatsteV1(Dataset):
    def __init__(self, txt, tokenizer, max_length, stride):
        self.input_ids = []
        self.target_ids = []

        #tokenize the entire text 
        token_ids = tokenizer.encode(txt , allowed_special={"<|endoftext|>"}) 

        for i in range(0 , len(token_ids) - max_length , stride):
            input_chunk = token_ids[i : i + max_length] #tokens from 0 to 4 
            target_chunk = token_ids[i + 1 : i + max_length + 1] # tokens from 1 to 5
            self.input_ids.append(torch.tensor(input_chunk))
            self.target_ids.append(torch.tensor(target_chunk))
    def __len__(self):
        return len(self.input_ids)  
    def __getitem__(self, idx):
        return self.input_ids[idx], self.target_ids[idx]
```

now for dataloader we begin with 
1 initialize the tokens 
2 create  dataset
3 drop last = true drops the last batch 
4 the number of cpu processes to use for preprocessing 

```python
def create_dataLoader_v1(txt, batch_size= 4, max_Length = 256 , stride = 128 , shuffle = True , drop_last = True , num_workers = 0):
    tokenizer = tiktoken.get_encoding("gpt2")
    dataset = GPTDatatsteV1(txt, tokenizer, max_Length, stride)

    DataLoader = DataLoader(dataset, batch_size=batch_size, shuffle=shuffle, drop_last=drop_last, num_workers=num_workers)
    return DataLoader
```
