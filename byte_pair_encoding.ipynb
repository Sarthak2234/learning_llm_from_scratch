```python
import importlib
import tiktoken 

print("tiktoken version:", importlib.metadata.version("tiktoken"))
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
#loads gpt2 tokenizer, which is used by many models including gpt-3.5-turbo and gpt-4
```

```python
text = (
    "Hello, do you like tea? <|endoftext|> In the sunlit terraces"
     "of someunknownPlace."
)
integers = tokenizer.encode(text, allowed_special={"<|endoftext|>"})
print(integers)
```

```python
strings = tokenizer.decode(integers)
print(strings)
```

here we can see how gpt2 enocder and decoder is working and how efficiently the unknown word ie someunknownPlace has been used.
